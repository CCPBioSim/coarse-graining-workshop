#### Part 1.3) Building the full membrane protein system using CHARMM-GUI

Navigate to CHARMM-GUI: https://www.charmm-gui.org/

Unfortunately, use of CHARMM-GUI requires an approved login account so we will be unable to walk you through its use in this tutorial.\
We point you to it here, however, as it is an invaluable resource in the construction and simulation of both AA and coarse grain (CG) systems.

Once logged in, there is a screen that shows links to all the tools available to build various types of molecular systems. In our case, \
we are interested in the Bilayer Builder subcomponent of the Membrane Builder toolset.

<img src="images/charmmgui-homescreen.png" width="750">

This tool takes you through various steps, including lipid selection, protein orientation and ion concentration specification, before\
finally providing a zipped file for download that contains a structure file for your model system, along with the topology, forcefield files\
and a generic equilibration protocol that allows you to perform a simulation "straight out of the box".

It is important, however, for you to independently verify that the system is correctly equilibrated at the end of this protocol, should you\
choose to use it, as it is in no way guaranteed to work for every unique system composition.
